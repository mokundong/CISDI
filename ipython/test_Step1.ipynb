```python
import pandas as pd
import numpy as np
res=pd.read_csv('res.csv')
import re
```

```python
rules = res.ix[:,1]
```

```python
len(rules)
```

```python
new_rules = []
reExp1 = r'X(.*?)=(.*?)[,}]'
```

```python
for tup in rules:
    invert = []
    invert = re.compile(reExp1).findall(tup)
    tmp = map(lambda x:int(x[0]) + int(x[1]),invert)
    tmp = list(tmp)
    tmp.sort()
    #print(tmp)
    new_rules.append(tmp)
```
